### 5.  Fit the model with the best subset of predictors (in terms of AIC). Report the fitted equation.

```python
from sklearn.metrics import mean_squared_error
# find the all AIC from all subset. 
list2 = ['lotsize','area','beds','baths','age','garage']
X = df[list2]
y = df['price']
```

```python
def get_sse(x, y):
    model = LinearRegression().fit(x, y)
    SSE = mean_squared_error(y, model.predict(x))*(len(y)-2)
    R_squared = model.score(x, y)
    return SSE, R_squared
```

```python
SSE_list, feature_list, num_features = [],[],[]
```

```python
df[list2].shape
```

```python
n = 522
p = 6
```

```python
import itertools
```

```python
m = len(X.columns)
list(range(1, m+1))
```

```python
for k in range(1, m+1):
    for subset in itertools.combinations(X.columns, k):
        temp = get_sse(X[list(subset)], y)
        SSE_list.append(temp[0])
        feature_list.append(subset)
        num_features.append(len(subset))
```

```python
df = pd.DataFrame(num_features, columns = ['n_features'])
df['feature'] = feature_list
SSE_list2 = np.round(SSE_list, 2)
df['SSE'] = SSE_list2
m = n
```

```python
df['AIC'] = n*np.log(df['SSE']/n) + 2*df['n_features']
df[:5]
```

```python
best_AIC = df.groupby(['n_features'])['AIC'].min()
df1 = df[df['AIC'] == best_AIC[1]]
```

```python
for i in range(2, 7):
    df1 = df1.append(df[df['AIC'] == best_AIC[i]])
df1
```

```python
row6 = df1[df1.AIC == df1.AIC.min()]
row6
# take this subset as model
```
